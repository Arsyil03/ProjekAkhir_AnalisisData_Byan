### Assessing Data

```python
zip_path = "data\\Air-quality-dataset.zip"
extract_path = "data\\Air-quality-extracted"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

data_folder = os.path.join(extract_path, "PRSA_Data_20130301-20170228")
df_air_quality = pd.concat(
    [
        pd.read_csv(os.path.join(data_folder, file)).assign(City=file.split("_")[2])
        for file in os.listdir(data_folder)
    ],
    ignore_index=True
)

df_air_quality.head()
```

## Exploratory Data Analysis (EDA)

### Explore Rata-rata Suhu per Kota 

```python
df_air_quality["TEMP"] = pd.to_numeric(df_air_quality["TEMP"], errors='coerce')

# Hitung rata-rata suhu per kota
city_temperature = df_air_quality.groupby("City")["TEMP"].mean().reset_index()

# Urutkan berdasarkan suhu rata-rata
city_temperature = city_temperature.sort_values(by="TEMP")

print("\nRata-rata Suhu per Kota, antara tahun 2013 - 2017 : ")
print(city_temperature)
```

```python
df_air_quality["PM2.5"] = pd.to_numeric(df_air_quality["PM2.5"], errors='coerce')

city_pollution = df_air_quality.groupby("City")["PM2.5"].sum().reset_index()
city_pollution = city_pollution.sort_values(by="PM2.5", ascending=False)

print("\nKota dengan Polutan Terbanyak (PM2.5) antara 2013-2017:")
print(city_pollution)
```

**Insight:**
Hasil analisis membuktikan jika Kota Huairou merupakan kota dengan Temperature terendah dibandingkan 12 kota lainnya di dalam dataset, sedangkan kota Gucheng merupakan kota dengan temperature tertinggi dibandingkan 12 kota lainnya. Hal ini di buktikan dari rata-rata suhu setiap kota nya dengan cara menghitung rata-rata suhu semua kota di rentang tahun 2013-2017 sesuai dengan dataset. 

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
df_air_quality["TEMP"] = pd.to_numeric(df_air_quality["TEMP"], errors='coerce')
df_air_quality["PM2.5"] = pd.to_numeric(df_air_quality["PM2.5"], errors='coerce')

city_temperature = df_air_quality.groupby("City")["TEMP"].mean().reset_index()
top_12_hottest = city_temperature.nlargest(12, "TEMP")  
top_12_coldest = city_temperature.nsmallest(12, "TEMP")  

city_pollution = df_air_quality.groupby("City")["PM2.5"].sum().reset_index()
city_pollution = city_pollution.sort_values(by="PM2.5", ascending=False)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
sns.barplot(
    y=top_12_hottest["City"], 
    x=top_12_hottest["TEMP"], 
    palette="Reds_r", 
    ax=axes[0]
)
axes[0].set_title("Top 12 Kota dengan Suhu Tertinggi Sepanjang Tahun", fontsize=12)
axes[0].set_xlabel("Suhu Maksimum (°C)")
axes[0].set_ylabel("Kota")
sns.barplot(
    y=top_12_coldest["City"], 
    x=top_12_coldest["TEMP"], 
    palette="Blues",  
    ax=axes[1]
)
axes[1].set_title("Top 12 Kota dengan Suhu Terendah Sepanjang Tahun", fontsize=12)
axes[1].set_xlabel("Suhu Minimum (°C)")
axes[1].set_ylabel("Kota")

plt.tight_layout()
plt.show()
```

### Pertanyaan 2:

```python
top_polluted_cities = city_pollution.head(12)

plt.figure(figsize=(8, 8))
plt.pie(
    top_polluted_cities["PM2.5"], 
    labels=top_polluted_cities["City"], 
    autopct="%1.1f%%", 
    colors=sns.color_palette("coolwarm", len(top_polluted_cities))
)
plt.title("Distribusi Polutan PM2.5 di 12 Kota dengan Polusi Tertinggi (2013-2017)")
plt.show()
```

**Insight:**
- Pada tahap Visualization pada pertanyaan pertama membuktikan bahwa kota dengan rata-rata suhu paling rendah terdapat pada kota Huairou.
- Pada tahap Visualization pada pertanyaan kedua membuktikan bahwa kota dengan rata-rata suhu paling tinggi terdapat pada kota Gucheng.
